# Importar as libraries necessárias

```python
#Fazer import dos pandas, das funções matemáticas, de uma extensão para fazer display de informação
#e do Numpy.
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display, HTML
import numpy as np

#Importar o csv para um DataFrame.
df = pd.read_csv("greenhouse.csv")

#Mostrar os primeiros 5 registos.
df.head()
```

```python
#Mostrar o tipo de dados em cada coluna da DataFrame
display(df.dtypes)
```

```python
#Descobrir quantos valores em falta existem
missing_values_count = df.isnull().sum()
missing_values_count
print("Existem ",missing_values_count.sum(), "valores em falta.")
```

```python
#Remover as linhas com células sem valores
df = df.dropna()
```

```python
#Mostrar quais os valores da coluna "magnitude"
mag_val = df.magnitude.unique()
display(mag_val)

#Descobrir quantos valores estão na coluna "magnitude" ao descobrir o comprimento da variável "mag_val".
num_mag = len(mag_val)
print("Na coluna display existem", num_mag, "valor único")
```

```python
#Remover colunas que são irrelevantes ou nao utilizamos
df = df.drop("magnitude", axis = 1, inplace = False)
df = df.drop("anzsic_descriptor", axis = 1, inplace = False)
df
```

# Quantidade de Gases expelidos por Região

### Calcular a quantidade total de gases expelidos em cada Região

```python
#Criar uma tabela, a partir da DataFrame inicial, apenas com a informação das Regiões e o 
#valor de gases em Kilotoneladas.
tabela_data = df[["region", "data_val"]]


#Alterar o Index da tabela para a coluna da Região.
tabela_data = tabela_data.set_index("region")
display(tabela_data)

#Cálculo da quantidade de gases por cada Região utilizando a função de soma, para somar todos 
#os registos com o mesmo nome de região.
total = tabela_data.loc[["Auckland"]].sum()
print("Total Data Value Auckland", total)

total = tabela_data.loc[["Bay of Plenty"]].sum()
print("Total Data Value Bay of Plenty", total)

total = tabela_data.loc[["Canterbury"]].sum()
print("Total Data Value Canterbury", total)

total = tabela_data.loc[["Gisborne"]].sum()
print("Total Data Value Gisborne", total)

total = tabela_data.loc[["Hawke's Bay"]].sum()
print("Total Data Value Hawke's Bay", total)

total = tabela_data.loc[["Manawatu-Whanganui"]].sum()
print("Total Data Value Manawatu-Whanganui", total)

total = tabela_data.loc[["Marlborough"]].sum()
print("Total Data Value Marlborough", total)

total = tabela_data.loc[["Northland"]].sum()
print("Total Data Value Northland", total)

total = tabela_data.loc[["Otago"]].sum()
print("Total Data Value Otago", total)

total = tabela_data.loc[["Southland"]].sum()
print("Total Data Value Southland", total)

total = tabela_data.loc[["Taranaki"]].sum()
print("Total Data Value Taranaki", total)

total = tabela_data.loc[["Tasman/Nelson"]].sum()
print("Total Data Value Tasman/Nelson", total)

total = tabela_data.loc[["Waikato"]].sum()
print("Total Data Value Waikato", total)

total = tabela_data.loc[["Wellington"]].sum()
print("Total Data Value Wellington", total)

total = tabela_data.loc[["West Coast"]].sum()
print("Total Data Value West Coast", total)
```

### Criar uma DataFrame com os valores obtidos na célula anterior 
### e com os nomes das Regiões

```python
#Criar a informação para inserir numa DataFrame.
#Criação de duas Colunas: Region e Total Data Value.
#Na coluna das Regiões, foram inseridas as 15 regiões mostrados na DataFrame inicial.
#Na outra coluna foram inseridos os valores obtidos na célula anterior em cada Região respetivamente.
data = {"Region": ["Auckland", "Bay of Plenty", "Canterbury", "Gisborne", "Hawke's Bay", "Manawater-Whanganui",
                   "Marlborough", "Northland", "Otago", "Southland", "Taranaki", "Tasman/Nelson", "Waikato", "Wellington",
                   "West Coast"],
           
        "Total Data Value": ["560865", "168122", "619053", "78597",  "182988", "330186", "43369", 
                             "273925", "274862", "330650", "318802", "62004", "769568", "179022", 
                             "72349"]
        }

#Criar a DataFrame com a informação anterior
df_value = pd.DataFrame (data, columns = ["Region","Total Data Value"])
#Alterar o Index da DataFrame para a coluna das Regiões
df_value = df_value.set_index("Region")
#Tornar todos os valores da Coluna "Total Data Value" para valores inteiros.
df_value["Total Data Value"] = df_value["Total Data Value"].astype(int)
#Mostrar a DataFrame
display(df_value)

#Calcular o total de gases de todas as regiões, somando todos os valores da Coluna "Total Data Value"
total_data = df_value["Total Data Value"].sum()
print("")
print("Total Data Value of Every Region:", total_data, "Killotons")
```

### Utilizar a DataFrame criada na célula anterior para dar origem a um gráfico de barras

```python
#Criação de um gráfico de barras com as quinze regiões no eixo x e a quantidade 
#de gases produzidos no eixo y, baseado na DataFrame anterior
plot2 = df_value.plot.bar(y = "Total Data Value", rot = 90, figsize=(20,10), fontsize = "15")
#Definir a label que aparece no eixo x
plot2.set_xlabel("Região", size=20)
#Definir a label que aparece no eixo y
plot2.set_ylabel("Gases Produzidos(Kilotoneladas)", size=20)
#Definir o titulo do gráfico
plot2.set_title("Quantidade de Gases Produzidos por Região em Kilotoneladas", size = 25)
#Desativar a legenda pois não é necessário para a compreensão do gráfico
plot2.legend(labels="")
```

### Utilizar a DataFrame criada anteriormente para dar origem a um gráfico circular

```python
#Definir uma lista com 15 valores HEX para aplicar às 15 regiões diferentes, respetivamente.
color = ["#FF4848", "#34D6FD", "#03EBA6", "#1FCB4A", "#2DC800", "#DB9900", "#FF8A1F", "#1F88A7",
         "#FF62B0", "#DD69FE", "#9669FE", "#800080", "#3923D6", "#2674AA", "#62A9FF"]

#Criação de um "Pie Chart" com as 15 regiões representadas a cores diferentes utilizando o 
#"Total Data Value" da DataFrame anterior
plot = df_value.plot.pie(y="Total Data Value", figsize=(10, 10), labels=None, colors = color)   
plot.set_title("Gases Produzidos por Região em Kilotoneladas", size = 25)
plot.set_ylabel(None)
#Alterar a deifinição da legenda do Gráfico. Nomeadamente a sua posição e a informação mostrada.
plot.legend(loc="lower right", labels=df_value.index)
```
